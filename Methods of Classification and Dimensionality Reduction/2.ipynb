* c) Simulate 500 points from $N(\boldsymbol{\mu},\boldsymbol{\Sigma})$, where
$$\boldsymbol{\mu}=(4,6,5)^T,
\boldsymbol{\Sigma}=\left(
  \begin{array}{llll}
  14 & 16 & 14 \\
  16 & 21 & 14 \\
  14 & 14 & 26
  \end{array}\right)$$
and draw them.
* d) Reduce the points to 2d via PCA. Draw these 2d points on a separate plot
* e) Reconstruction: using PCA matrix, convert these 2d points back to 3d, plot them together with original points (with different color)

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Plot 3D PCA
fig = plt.figure(1)
ax = fig.add_subplot(111, projection='3d')

# Scatter plot points, differentiating species by color
for species in range(3):
    indices = y == species
    ax.scatter(X_pca[indices, 0], X_pca[indices, 1], X_pca[indices, 2], label=iris.target_names[species])

ax.set_title("IRIS PCA 3D")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.legend()

# Plot 2D PCA
fig, ax = plt.subplots(figsize=(6, 4))

# Scatter plot points, differentiating species by color
for species in range(3):
    indices = y == species
    ax.scatter(X_pca[indices, 0], X_pca[indices, 1], label=iris.target_names[species])

ax.set_title("IRIS PCA 2D")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.legend()

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import fetch_openml


mnist = fetch_openml('mnist_784', version=1)
X = mnist.data.astype('float32')
y = mnist.target.astype('int')

# Normalize the data - normalizing the pixel values to the range [0, 1] helps ensure that each pixel contributes proportionally to the overall representation of the digit images
X /= 255.0


pca = PCA(n_components=3)
X_pca_3d = pca.fit_transform(X)

pca = PCA(n_components=2)
X_pca_2d = pca.fit_transform(X)

# Plot 2D PCA
plt.figure(figsize=(10, 6))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap=plt.cm.get_cmap('tab10', 10))
plt.colorbar(ticks=range(10), label='Digit')
plt.title('MNIST PCA 2D Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Plot 3D PCA
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], c=y, cmap=plt.cm.get_cmap('tab10', 10))
legend = ax.legend(scatter.legend_elements(), title='Digit')
ax.add_artist(legend)
ax.set_title('MNIST PCA 3D Visualization')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = np.array([4, 6, 5])
cov = np.array([[14, 16, 14],
                [16, 21, 14],
                [14, 14, 26]])

# Generate 500 points
points = np.random.multivariate_normal(mu, cov, size=500)

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', marker='o', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('500 Points from Multivariate Normal Distribution')
plt.show()
````

````python
pca = PCA(n_components=2)
points_2d = pca.fit_transform(points)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(points_2d[:, 0], points_2d[:, 1], c='b', marker='o', alpha=0.5)
plt.title('500 Points from Multivariate Normal Distribution (2D PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


mu = np.array([4, 6, 5])
cov = np.array([[14, 16, 14],
                [16, 21, 14],
                [14, 14, 26]])


points = np.random.multivariate_normal(mu, cov, size=500)


pca = PCA(n_components=2)
points_2d = pca.fit_transform(points)

# Reconstruct 2D points back to 3D
reconstructed_points = pca.inverse_transform(points_2d)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', marker='o', alpha=0.5, label='Original Points')


ax.scatter(reconstructed_points[:, 0], reconstructed_points[:, 1], reconstructed_points[:, 2], c='r', marker='o', alpha=0.5, label='Reconstructed Points')

ax.set_title('Original and Reconstructed 3D Points')

ax.legend()

plt.show()
````
